*A* notebook consists of cells in which users are able to type text. There exist two different types of cells: Code and Text. As its name suggests, a code cell accepts Python statements. On the other hand, a text cell enables you to enter ordinary and Markdown-formatted texts.

## Kernel

A kernel is a “computational engine” that executes the code contained in a notebook document. Whenever you run a code cell, the content of the cell is executed by the kernel. In addition, the defined variables, function, class, etc. are saved in the memory so that you can use them in another code cell. Each time you restart your kernel, all the definitions will be gone.

```python
# a variable defined here
x = 5
```

```python
# can be used in another code cell as well
print(x) 
```

*In contrast to the conventional computation environments, such as IDEs, the execution order is not linear. You may define a variable in the lower parts of the notebook and still access it. However, this is not good practice. Next time you run the notebook from the top, you may have `"name ... is not defined"` errors.*

## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.show()
```

## Mounting Your Drive

As stated in the section above, your access to a notebook is volatile. The next time you launch this notebook, your file(s) may not be there. In order to prevent this problem, you may utilize your drive and use the space there.

To access the files in your drive, you need to authorize your Colab notebook. To do that, the code snippet below must be executed. 

Once you execute the cell, a link will appear.  

- Click on it and follow the instructions. 
- In the end, an authorization code is generated. 
- Copy/Paste it to the blank area and hit Enter.

Now, you should be able to access your drive. A new folder named "drive" is created in which your entire drive is contained in another folder named "My Drive". 

```python
# execute this cell to authorize the notebook
from google.colab import drive
drive.mount("./drive", force_remount=True)
```
